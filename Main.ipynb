```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
df = pd.read_csv('bank-full.csv', sep=';')
df.head(15)
```

```python
df['y'] = df['y'].replace({'yes': 1, 'no': 0})
```

```python
# Podzia≈Ç kolumn wg. typu danych (numeryczne)

numeric_columns = df[['age', 'balance', 'day', 'duration', 'campaign', 'pdays', 'previous']]
categorical_columns = df[['job', 'marital', 'education', 'contact', 'month', 'poutcome']]
binary_columns = df[['default', 'housing', 'loan', 'y']]
```

```python
df.nunique()
```

```python
df.isnull().sum()/len(df)
```

```python
# Policz liczebno≈õƒá kolumny 'y'
y_count = df['y'].value_counts()
sd= y_count.index
counts = y_count.values


# Tworzenie wykresu
plt.figure(figsize=(10, 6))
bars = plt.bar(y_count.index, counts, color="skyblue", edgecolor="black")
plt.title("Rozk≈Çad sukses√≥w kampanii", fontsize=14, fontweight='bold')
plt.xlabel("y", fontsize=12)
plt.ylabel("count", fontsize=12)

# Dodanie etykiet
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, height + 0.5, f'{int(height)}',
             ha='center', va='bottom', fontsize=10)

plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.tight_layout()

```

```python
for col in numeric_columns:   
    sns.histplot(df[col], bins = 50, kde = True),
    plt.title(f'Dystrybucja {col}')
    plt.show()
```

```python
for col in categorical_columns:   
    sns.countplot(x=df[col], order = df[col].value_counts().index)
    plt.title(f'Count of {col}')
    plt.xticks(rotation=90)
    plt.show()
```

```python
for col in binary_columns:   
    sns.countplot(x=df[col], order = df[col].value_counts().index)
    plt.grid(True, axis='y', alpha=0.8, linestyle='--', linewidth=0.5, color='gray')
    plt.title(f'Count of {col}')
    plt.xticks()
    plt.show()
```
