```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
data = pd.read_csv('bank-full.csv', sep=';')
```

```python
plt.figure(figsize=(8, 5))
plt.hist(
    data["age"], 
    bins=30, 
    color="cornflowerblue", 
    edgecolor="black",
    alpha=0.8
)

# Dodanie tytułu i etykiet
plt.title("Participants by Age", fontsize=14, fontweight='bold')
plt.xlabel("Age", fontsize=12)
plt.ylabel("Count", fontsize=12)

# Dodanie siatki i dopasowanie wyglądu
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.tight_layout()

# Zapis wykresu do pliku PNG

# Wyświetlenie wykresu
plt.show()
plt.close()
```

```python
plt.figure(figsize=(8, 5))
sns.kdeplot(
    data=data,
    x="balance",
    fill=True,              # wypełnienie pod krzywą
    color="cornflowerblue",
    alpha=0.5,
    linewidth=2
)

# Dodanie tytułu i etykiet
plt.title("Rozkład gęstości wieku uczestników", fontsize=14, fontweight='bold')
plt.xlabel("Wiek", fontsize=12)
plt.ylabel("Gęstość", fontsize=12)

# Siatka i wygląd
plt.grid(axis='y', linestyle='--', alpha=0.5)

# Zapis wykresu do pliku PNG

# Wyświetlenie wykresu
plt.show()
plt.close()
```
