```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
plt.figure(figsize=(8, 5))
plt.hist(
    df["age"], 
    bins=30, 
    color="cornflowerblue", 
    edgecolor="black",
    alpha=0.8
)

# Dodanie tytuu i etykiet
plt.title("Participants by Age", fontsize=14, fontweight='bold')
plt.xlabel("Age", fontsize=12)
plt.ylabel("Count", fontsize=12)

# Dodanie siatki i dopasowanie wygldu
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.tight_layout()

# Zapis wykresu do pliku PNG

# Wywietlenie wykresu
plt.show()
plt.close()
```

```python
plt.figure(figsize=(8, 5))
plt.hist(
    df["duration"], 
    bins=30, 
    color="cornflowerblue", 
    edgecolor="black",
    alpha=0.8
)

# Dodanie tytuu i etykiet
plt.title("Duration of campaign", fontsize=14, fontweight='bold')
plt.xlabel("Duration", fontsize=12)
plt.ylabel("Count", fontsize=12)

# Dodanie siatki i dopasowanie wygldu
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.tight_layout()

# Zapis wykresu do pliku PNG

# Wywietlenie wykresu
plt.show()
plt.close()
```
